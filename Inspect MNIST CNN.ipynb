```python
batch_size = 500
valid_data = datasets.MNIST(root='./data', train=False, 
                            download=True, transform=transforms.ToTensor())
valid_loader = torch.utils.data.DataLoader(valid_data, shuffle=False, batch_size=batch_size)

plt.gray()

Xb, yb = next(iter(valid_loader))
yb_pred = net(Xb).argmax(dim=1)

show_image_grid(
    Xb, yb_pred, y_true=yb,
    nrow=10, ncol=50, sz_factor=1.5
)
```
